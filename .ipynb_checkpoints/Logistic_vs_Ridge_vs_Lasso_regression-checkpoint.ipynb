```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
import seaborn as sns
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn import linear_model
from sklearn import preprocessing
%matplotlib inline
sns.set_style('white')
```

```python
df=pd.read_excel(io='table_8_offenses_known_to_law_enforcement_california_by_city_2013.xls')
```

```python
df=df.drop(df.index[:4])
```

```python
df=df.drop(df.index[348:])
```

```python
df.columns=['City','Population','Violent crime','Murder and nonnegligent manslaughter','Rape (revised definition)','Rape (legacy definition)','Robbery','Aggravated assault','Property crime','Burglary','Larceny- theft','Motor vehicle theft','Arson']
```

```python
df['Population']=pd.to_numeric(df['Population'], errors='coerce')
df['Violent crime']=pd.to_numeric(df['Violent crime'], errors='coerce')
df['Murder and nonnegligent manslaughter']=pd.to_numeric(df['Murder and nonnegligent manslaughter'], errors='coerce')
df['Rape (revised definition)']=pd.to_numeric(df['Rape (revised definition)'], errors='coerce')
df['Rape (legacy definition)']=pd.to_numeric(df['Rape (legacy definition)'], errors='coerce')
df['Robbery']=pd.to_numeric(df['Robbery'], errors='coerce')
df['Aggravated assault']=pd.to_numeric(df['Aggravated assault'], errors='coerce')
df['Property crime']=pd.to_numeric(df['Property crime'], errors='coerce')
df['Burglary']=pd.to_numeric(df['Burglary'], errors='coerce')
df['Larceny- theft']=pd.to_numeric(df['Larceny- theft'], errors='coerce')
df['Motor vehicle theft']=pd.to_numeric(df['Motor vehicle theft'], errors='coerce')
df['Arson']=pd.to_numeric(df['Arson'], errors='coerce')
```

```python
df.head()
```

```python
sns.boxplot(x="Violent crime", data=df)
plt.show()
```

```python
df['Violent crime'].describe()
```

```python
# Create a binary indicator of violent crime based on the mean of the data set
features = df.copy()
features['Violent_crime_indicator'] = np.where(features['Violent crime'] > 248.262494, 1, 0)
```

```python
features.tail()
```

```python
features=features.drop(['Violent crime'],axis=1)
```

```python
features.head()
```

```python
trainsize=int(features.shape[0]/2)
features_test = features.iloc[trainsize:, :].copy()
features_train = features.iloc[:trainsize,:].copy()
```

```python
# Declare a logistic regression classifier.
# Parameter regularization coefficient C described above.
lr = LogisticRegression(C=1e9)
Y_train = features_train['Violent_crime_indicator']
X_train = features_train[['Population', 'Murder and nonnegligent manslaughter','Rape (legacy definition)','Robbery','Aggravated assault','Property crime','Burglary','Larceny- theft','Motor vehicle theft','Arson']]

# Fit the model.
fit = lr.fit(X_train, Y_train)

# Display.
print('Coefficients')
print(fit.coef_)
print(fit.intercept_)
pred_y_sklearn = lr.predict(X_train)

print('\n Accuracy by admission status')
print(pd.crosstab(pred_y_sklearn, Y_train))

print('\n Percentage accuracy')
print(lr.score(X_train, Y_train))
```

```python
ridgeregr = linear_model.Ridge(alpha=10, fit_intercept=False) 
ridgeregr.fit(X_train, Y_train)
print('R²:')
print(ridgeregr.score(X_train, Y_train))
origparams = ridgeregr.coef_[0]
print('\nParameter estimates:')
print(origparams)

# Display.
print('Coefficients')
print(ridgeregr.coef_)
print(ridgeregr.intercept_)
pred_y_sklearn = ridgeregr.predict(X_train)

print('\n Accuracy by admission status')
print(pd.crosstab(pred_y_sklearn, Y_train))

print('\n Percentage accuracy')
print(lr.score(X_train, Y_train))
```

```python
# Small number of parameters.
lass = linear_model.Lasso(alpha=.35)
lassfit = lass.fit(X_train, Y_train)
print('R²:')
print(lass.score(X_train, Y_train))
origparams = np.append(lassfit.coef_, lassfit.intercept_)
print('\nParameter estimates:')
print(origparams)

# Display.
print('Coefficients')
print(lass.coef_)
print(lass.intercept_)
pred_y_sklearn = lass.predict(X_train)

print('\n Accuracy by admission status')
print(pd.crosstab(pred_y_sklearn, Y_train))

print('\n Percentage accuracy')
print(lr.score(X_train, Y_train))
```
